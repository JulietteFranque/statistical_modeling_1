```python
from scipy.stats import norm, lognorm
import numpy as np
```

```python
c = 2
mu = 10
dist = norm(mu, np.sqrt(mu**2 * 5))
```

```python
sample = dist.rvs(1000)
```

```python
dist.pdf(sample)
```

```python
prods = []
mus = np.linspace(1, 100, 1000)
for mu in mus:
    f = norm(mu, mu**2 * 5).logpdf(sample)
    prods.append(np.sum(f))
```

```python
import matplotlib.pyplot as plt
plt.plot(mus, prods)
```

```python
mus[np.argmax(prods)]
```
